# bank-full Dataset

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
bank=pd.read_csv('bank-full.csv',sep=";")
bank
```

```python
bank.info()
```

```python
bank.head(12)
```

## EDA

```python
# creating dummy columns for the categorical columns 
data1=pd.get_dummies(bank,columns=['job','marital','education','contact','month','poutcome'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
# Custom Binary Encoding 
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

## Select features and target¶

```python
# Dividing our data into features(input) and target(output) variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

## Model creation

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

## Prediction¶

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

## Accuracy testing (confusion matrix)¶

```python
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39162+1153)/(39162+760+4136+1153)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
# As accuracy = 0.8917, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

## ROC-AUC curve¶

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
